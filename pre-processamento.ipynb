```python
df11 = df1.dropna()
df11.shape
```

```python
print(df11.isnull().sum())
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df111 = scaler.fit_transform(df11)
df1_norm = pd.DataFrame(data=df111, columns=df11.columns)
df1_norm.head()
```

```python
df1_norm['class'] = df1_norm['class'].astype('category')
df1_norm['class'] = df1_norm['class'].cat.codes
df1_norm.head()
```

```python
df1_norm[df1_norm['class']==1].head()
```

```python
df_shuffle = df1_norm.sample(frac = 1, random_state=42)
df_shuffle.head()
```

```python
df_shuffle.drop_duplicates(inplace=True)
df_shuffle.shape
```

```python
df_shuffle['class'].value_counts()
```

```python
df_shuffle.to_csv('clean_data.csv')
```
